```python
#prepaing dataset
df1=df.iloc[:,1:]
df1
```

```python
#splitting data for test and train
from sklearn.model_selection import train_test_split

# Assuming 'Type' is your target variable
X = df1.drop('type', axis=1)
Y = df1['type']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

# Now you have X_train, X_test, y_train, and y_test for model training and evaluation
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# Model Building

```python
# KNN Classification
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
```

```python
#parameters
#num_folds = 10
kfold = KFold(n_splits=5)
knn_model = KNeighborsClassifier(n_neighbors=3)
scores = cross_val_score(knn_model, X, Y, cv=kfold)
```

```python
# Print the results
print(f"Cross-Validation Accuracy Scores: {scores}")
print(f"Mean Accuracy: {scores.mean():.2f}")
print(f"Standard Deviation: {scores.std():.2f}")
```

```python
# Train the model
knn_model.fit(X_train, y_train)

# Predict on the test set
y_pred = knn_model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

# **Grid Search for Algorithm Tuning**

```python
#search for best parameters
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_neighbors': range(1, 20),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
```

```python
#model building
from sklearn.neighbors import KNeighborsClassifier

knn_best = KNeighborsClassifier(metric='manhattan', n_neighbors=1, weights='uniform')
knn_best.fit(X_train, y_train)
```

```python
#testing accuracy
test_accuracy = knn_best.score(X_test, y_test)
print("Test Accuracy with Best Parameters:", test_accuracy)
```

```python
#results
from sklearn.metrics import classification_report, confusion_matrix

y_pred = knn_best.predict(X_test)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# Visualizing the CV results

```python
# cv plot
k_range = range(1, 41)
k_scores = []
# use iteration to caclulator different k in models, then return the average accuracy based on the cross validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=5)
    k_scores.append(scores.mean())
# plot to see clearly
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

# **Interview Questions:**

```python
#1. What are the key hyperparameters in KNN?
'''
n_neighbors: The number of nearest neighbors to consider when making a prediction.
metric: The distance metric used to calculate similarity (e.g., Euclidean, Manhattan).
weights: Determines how neighbors influence predictions (e.g., uniform or distance).
algorithm: The method for finding nearest neighbors (e.g., auto, kd_tree, ball_tree, brute).
leaf_size: The size of leaf nodes in tree-based algorithms, affecting speed and memory.
p: The power parameter for the Minkowski distance, generalizing Euclidean and Manhattan distances.
'''

#2. What distance metrics can be used in KNN?
'''
The distance metrics used in KNN are:

Euclidean: The straight-line distance between two points (default).
Manhattan: The sum of absolute differences along each dimension.
Minkowski: A generalization of Euclidean and Manhattan (parameterized by p).
Cosine: Measures the cosine of the angle between two vectors (used in text classification).
Hamming: Measures the number of differing bits between two binary vectors.
Chebyshev: The maximum absolute difference along any dimension.
'''
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap # Import ListedColormap
import numpy as np


# Reduce the data to 2D using PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_train)

# Train KNN on the reduced data
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_reduced, y_train)

# Create a meshgrid for the reduced data
x_min, x_max = X_reduced[:, 0].min() - 1, X_reduced[:, 0].max() + 1
y_min, y_max = X_reduced[:, 1].min() - 1, X_reduced[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Predict the class for each point in the grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundaries
plt.contourf(xx, yy, Z, alpha=0.75, cmap=ListedColormap(("red", "green", "blue"))) # Now ListedColormap is recognized
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_train, cmap=ListedColormap(("red", "green", "blue")), edgecolor='k', s=50)
plt.title("KNN Decision Boundary with PCA Reduced Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
